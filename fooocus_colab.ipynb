```python
# Step 1: Define a dictionary of FAQs
faq_data = {
    "what is python": "Python is a high-level programming language used for web development, data science, AI, and more.",
    "what is ai": "AI (Artificial Intelligence) is the simulation of human intelligence in machines that can learn and make decisions.",
    "what is machine learning": "Machine Learning is a subset of AI that allows systems to learn from data and improve over time.",
    "what is google colab": "Google Colab is a free cloud-based platform that lets you run Python code in a Jupyter Notebook environment.",
    "what is a database": "A database is an organized collection of data that can be accessed, managed, and updated easily.",
    "what is an algorithm": "An algorithm is a step-by-step procedure or formula for solving a problem."
}

# Step 2: Create a chatbot function
def chatbot(query):
    query = query.lower().strip()  # normalize input
    if query in faq_data:
        return faq_data[query]
    else:
        return "Sorry, I don't know the answer to that. Please try asking something else."

# Step 3: Interactive loop
print("Welcome to the Academic FAQ Chatbot! Type 'exit' to quit.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye! Have a great day.")
        break
    response = chatbot(user_input)
    print("Chatbot:", response)
```
